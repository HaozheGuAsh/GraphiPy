```python
from usde.usde import USDE

# Youtube Api's credentials
api_credentials = {'api_key': ' ',	
           'client_secret': "client_secrets.json"}

# Create usde
usde = USDE()
# Create YouTube
youtube = usde.get_youtube(api_credentials)
```

# Searching for videos with topic "USC"

```python
# Create a graph and naming the graph type inside usde first and then modify the graph
graph = usde.create_graph("pandas")
youtube.fetch_videos_by_topic(graph, "usc")
nodes = graph.get_nodes()
df = nodes["video"].head(10)
```

# Let's search for one of the videos and get its comment

```python
info_of_one_video_graph = usde.create_graph("pandas")
# let's search for "USC vs. Colorado | FOX COLLEGE FOOTBALL HIGHLI..." which has id of DAV2sWfiZN4
youtube.fetch_video_by_id_with_comments(info_of_one_video_graph, "DAV2sWfiZN4")
nodes = info_of_one_video_graph.get_nodes()
df = nodes["comment"].head(20)
```

# For all the USC video, we can also find the channels and the edge between channel and video

```python
graph = usde.create_graph("pandas")
youtube.fetch_videos_by_topic(graph, "usc")
nodes = graph.get_nodes()
df = nodes["channel"].head(10)
```

```python
edges = graph.get_edges()
df = edges["createvideo"].head(25)
```
