```python
tumblr = usde.get_tumblr(tumblr_api_credentials)
```

# Find posts by a tag

```python
# Every function call modifies the graph that is sent as input
# tagged_posts = usde.create_graph()
# tumblr.fetch_posts_tagged(graph=tagged_posts, tag="usc", limit=5)

# However, it also returns the graph modified so you can assign it to other variables like so:
tagged_posts = tumblr.fetch_posts_tagged(graph=usde.create_graph(), tag="usc", limit=5)
```

```python
# You can also get the dataframe from Pandas
tagged_posts_df = tagged_posts.get_df("post")
```
