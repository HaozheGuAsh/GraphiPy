# Visualize Neo4j Graph Using NetworkX

```python
from graphipy.graphipy import GraphiPy
```

```python
# create GraphiPy object
graphipy = GraphiPy("neo4j")
```

```python
# The reddit API needs these credentials
CLIENT_ID = ""
CLIENT_SECRET = ""
USER_AGENT = ""
USERNAME = ""
PASSWORD = ""
reddit_api_credentials = {
    "client_id": CLIENT_ID,
    "client_secret": CLIENT_SECRET,
    "user_agent": USER_AGENT,
    "username": USERNAME,
    "password": PASSWORD
}

# create the reddit object
reddit = graphipy.get_reddit(reddit_api_credentials)
```

# Find Comments in a Submission

```python
# Sample query
neo4j_auth = ""
graph = graphipy.create_graph("neo4j", neo4j_auth)

submission_id = "9uolpb"
graph = reddit.fetch_submission_comments(graph, submission_id, limit=20)
```

# Visualization with NetworkX

```python
# Create graph from Neo4j
nodes = graph.get_nodes()
edges = graph.get_edges()
nx_graph = graphipy.nx_create_from_neo4j(nodes, edges)
```

```python
# Draw the graph
options = {
    "node_label": "Label",
    "colorful_edges": True
}
legend = graphipy.nx_draw_random(nx_graph, options=options, legend=True)
```
