# Using CSV Files with NetworkX

```python
from graphipy.graphipy import GraphiPy
```

```python
graphipy = GraphiPy()
```

```python
# Create graph from csv files

path_to_edges = ""
path_to_nodes = ""

# For a directed graph, set directional=True when calling the function
# However, edges may overlap due to networkx's limitations
nx_graph = graphipy.nx_create_from_csv(path_to_edges, path_to_nodes)
```

```python
# Draw the graph
options = {
    "node_label": "Label",
    "colorful_edges": True
}
legend = graphipy.nx_draw_random(nx_graph, options = options, legend=True)
```
